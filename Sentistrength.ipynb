```
for ind in df.index:
  positive = int(float(df['Positive'][ind]))
  neg = df['Negative'][ind]
  negative = float(df['Negative'][ind])
  if(positive == abs(negative)):
    df['Positive'][ind] = "Neutral"
  elif(positive > abs(negative)):
    df['Positive'][ind] = "Positve"
  else:
    df['Positive'][ind] = "Negative"
```
