```
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
df = pd.read_csv('drive/My Drive/Data Mining Spring 2020/us_correlation_analysis.csv')
df
```

```
df_sentiment_cases = df[['sentiment_us', 'cases_daily']].copy()
df_sentiment_tweets = df[['sentiment_us', 'tweets_daily']].copy()
df_tweets_cases = df[['tweets_daily', 'cases_daily']].copy()
```

```
df_tweets_cases
```

```
# Linear regression for sentiments v/s cases daily
Y = df_sentiment_cases.iloc[:, 0].values.reshape(-1, 1)  
X = df_sentiment_cases.iloc[:, 1].values.reshape(-1, 1)  
```

```
# perform linear regression
linear_regressor = LinearRegression()  
linear_regressor.fit(X, Y)  
Y_pred = linear_regressor.predict(X) 
```

```
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```
