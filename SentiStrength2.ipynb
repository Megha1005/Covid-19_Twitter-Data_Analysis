```
df["Sentiment"] = ""
```

```
for ind in df.index:
  positive = int(float(df['Positive'][ind]))
  neg = df['Negative'][ind]
  negative = float(df['Negative'][ind])
  if(positive == abs(negative)):
    df['Sentiment'][ind] = "Neutral"
  elif(positive > abs(negative)):
    df['Sentiment'][ind] = "Positve"
  else:
    df['Sentiment'][ind] = "Negative"
```
